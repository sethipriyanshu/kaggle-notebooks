# The Curse of <mark> Dimensionality</mark>
<hr>

<img src="https://www.shiksha.com/online-courses/articles/wp-content/uploads/sites/11/2022/05/curse-of-dimensionality.png">

<p style="font-size:130%;"> Every Feature (column) of a given Dataset is called a dimension. There is always an optimal number of dimensions beyond which any new features do no positively contribute to our model. In Fact , This reduces the accuracy and increases the run time while training our model by making it more computationaly complex.</p><hr>

### Let's Understand this with a basic example



<p style="font-size:130%;">Suppose We have a dataset with the RGB Value of each pixel as and individual column. The first approach could be using the first image given above as a whole but notice that most of the pixels are white in color and have no meaningful contribution in the quality of our model.<br>Now,We might decide to not use all the columns and for better understanding let's assume that the pixels shaded black are ignored. What We'll observe is that the second image gives us equally as good as accuracy while using less computational power.<br> Again, We could make a mistake and choose the third image which would adversely affect our model but feature selection is a whole different thing to discuss. Let's focus on Feature extraction for now</p><hr>

# How does PCA work?
<hr>

<img src="https://i.ytimg.com/vi/XxeVsGjvUxQ/maxresdefault.jpg">
<p style="font-size:130%;">Let's use the analogy of a soccer match to depict how PCA works. As we all know , Soccer matches are held in a 3-D environment and the goal of the cameraman is to record the same match and telecast it in a 2-D format. Everything about the match remains same and just the dimensionality reduces. Similarly , PCA is used to bring a high dimensionanal dataset to the best possible low dimension form by using existing features to extract new ones</p><hr>

### Benefits of using PCA
<p style="font-size:130%;">-Faster Execution of Algorithms <br>-Easier Visualization<br>-Helps Overcome overfitting issues.</p><hr>
